## move some images to test path for testing 

```python
import os
import glob as gb
import random
import shutil

dataset = 'Dataset'
trainpath = 'HMBD v1 dataset/trainpath'
testpath = 'HMBD v1 dataset/testpath'
predpath = 'HMBD v1 dataset/predpath'

for folder in os.listdir(dataset):
    files = gb.glob(pathname=str(dataset + '/' + folder + '/*.jpg'))
   
    randomNO = random.randint(100, 115)
    random_images = random.sample(files, randomNO)
    print(len(random_images))
 
    for img_path in random_images:
        file_name = os.path.basename(img_path)
        destination_folder = os.path.join(testpath, folder)
        os.makedirs(destination_folder, exist_ok=True)
        destination_path=os.path.join(destination_folder, file_name)
        shutil.move(img_path,destination_path)        
```

### show count image in all folder in testpath

```python
total=0
for folder in  os.listdir(testpath ) : 
    files = gb.glob(pathname= str( testpath +'/'+ folder + '/*.jpg'))
    total+=len(files)
    print(f' found {len(files)} files in folder {folder}')
    
print('='*50)
print(f'total count files in {testpath} is : {total}')
print('='*50)
```
